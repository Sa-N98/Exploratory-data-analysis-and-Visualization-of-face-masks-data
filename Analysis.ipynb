```python
# Creating New Data Frame By Expamding product_name:
data=[]
for i in product['product_name']:
    lis=i.split(',')
    print(i, len(lis))
    if len(lis)== 3:
        data.append([lis[0],lis[1],None,None,lis[-1].split(" ")[-2]])
    elif len(lis)== 4:
        data.append([lis[0],lis[1],lis[2],None,lis[-1].split(" ")[-2]])
    elif len(lis)== 5:
        data.append([lis[0],lis[1],lis[2],lis[3],lis[-1].split(" ")[-2]])

df=pd.DataFrame(data, columns=["Mask_Name", "Type", "Size","Design","Pack"])


# Fixing Miss Labeled Data: 
maks_type= df.loc[2]["Size"]+ df.loc[2]["Type"]
df.loc[2][['Type','Size']]=[maks_type,None]
df.loc[21][['Size', "Design"]]=['S_M','Grey']


# Adding product_id and Product_price To New Data Frame: 
df['product_id']= product['product_id']
df['product_price'] = product['product_price']
```
